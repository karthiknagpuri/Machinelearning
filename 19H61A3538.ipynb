```python
a=np.random.rand(2,2)
a1=np.linalg.inv(a)
a2=a@a1
print(a)
print(" ")
print(a1)
print(" ")
print(a2)
```

```python
fig,x=plt.subplots(1,3,figsize=(6,8))
x[0].imshow(a)
x[0].set_title('a')
x[1].imshow(a1)
x[1].set_title('a$^{-1}$')
x[2].imshow(a2)
x[2].set_title('a$^{-1}$a')
plt.show()
```
